```python
x= df.drop('prognosis',axis='columns')
y=df['prognosis']
```

```python
seen = set()
unique_elements = []
a=0
for element in y:
    if element not in seen:
        unique_elements.append(element)
        seen.add(element)

# Print the unique elements
for i,element in enumerate(unique_elements):
    print(f"symptom number {i}    {element}")
    
    
```

```python
y_dash= pd.factorize(y)[0]
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y_dash,test_size=0.2,random_state=5)
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix


```

```python

model = RandomForestClassifier()
model.fit(xtrain, ytrain)
```

```python
ypred = model.predict(xtest)
```

```python
import seaborn as sn
cm = tf.math.confusion_matrix(labels=ytest,predictions=ypred)

plt.figure(figsize = (10,7))
sn.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
def predict_disease(symptoms_input):
    # Assuming `symptoms_list` contains the column names of your symptoms
    symptoms_list = x.columns
    
    # Create an input vector for the model
    input_vector = np.zeros(len(symptoms_list))
    
    # Set the symptoms provided by the user to 1
    for symptom in symptoms_input:
        if symptom in symptoms_list:
            index = symptoms_list.get_loc(symptom)
            input_vector[index] = 1
    
    # Predict the disease
    prediction = model.predict([input_vector])[0]
    return prediction
```
